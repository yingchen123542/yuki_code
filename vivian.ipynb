```python
#youtube word count
index=0
#initialize lists to store word count  
youtube_word = set()
setlist_youtube=[]
words_list_yt=[]
count_yt=0
count_list_yt=[]
#remove punctuation from sentences 
for cell in youtube:
    youtube=str(cell)
    youtube=youtube.strip('!?.,:"')
    #convert sentence into lower case 
    youtube=youtube.lower()
    #split sentence into words in order to count the number of words 
    sentence_list=youtube.split() 
    words_list_yt.append(youtube)
#loop through the list to count the number of words 
for sentence in words_list_yt:
    word_yt=sentence.split()
    if(word_yt[0]=="nan"):
        count_yt=0
    else:
        count_yt=len(word_yt)
    
    count_list_yt.append(count_yt)
    for element in word_yt:
        #add words into set to calculate symmetric difference 
        youtube_word.add(element)
    setlist_youtube.append(youtube_word)
    youtube_word = set()

print("youtube word count:",count_list_yt)
print(setlist_youtube)
```

```python
def wer1(r, h):
    # initialisation
    import numpy
    d = numpy.zeros((len(r)+1)*(len(h)+1), dtype=numpy.uint8)
    d = d.reshape((len(r)+1, len(h)+1))

    for i in range(len(r)+1):
        for j in range(len(h)+1):
            if i == 0:
                d[0][j] = j
            elif j == 0:
                d[i][0] = i
    # computation
    for i in range(1, len(r)+1):
        for j in range(1, len(h)+1):
            if r[i-1] == h[j-1]:
                d[i][j] = d[i-1][j-1]
            else:
                substitution = d[i-1][j-1] + 1
                insertion    = d[i][j-1] + 1
                deletion     = d[i-1][j] + 1
                d[i][j] = min(substitution, insertion, deletion)
     
    i = len(r)
    j = len(h)
    
    insert = 0
    delete = 0
    match = 0
    mismatch =0
    
    while (i>0 or j>0):
        if (d[i-1][j-1]<d[i-1][j] and d[i-1][j-1]<d[i][j-1]):
            if r[i-1] == h[j-1]:
                match += 1
            else:
                mismatch +=1               
            i = i-1
            j = j-1
       
        elif (d[i-1][j] < d[i][j-1]):
            i=i-1
            delete = delete +1
        else:
            insert = insert + 1
            j=j-1
    print ("match ", match)
    match_list_yt.append(match)
    print ("mismatch ", mismatch)
    mismatch_list_yt.append(mismatch)
    print ("delete", delete)
    delete_list_yt.append(delete)
    print ("insert", insert)
    insert_list_yt.append(insert)
    return insert
```

```python
#initialize list to store word error rate for ibm vs ground true 
word_error_rate_list_ibm=[]

def wer1(r, h):
    # initialisation
    import numpy
    d = numpy.zeros((len(r)+1)*(len(h)+1), dtype=numpy.uint8)
    d = d.reshape((len(r)+1, len(h)+1))

    for i in range(len(r)+1):
        for j in range(len(h)+1):
            if i == 0:
                d[0][j] = j
            elif j == 0:
                d[i][0] = i
    # computation
    for i in range(1, len(r)+1):
        for j in range(1, len(h)+1):
            if r[i-1] == h[j-1]:
                d[i][j] = d[i-1][j-1]
            else:
                substitution = d[i-1][j-1] + 1
                insertion    = d[i][j-1] + 1
                deletion     = d[i-1][j] + 1
                d[i][j] = min(substitution, insertion, deletion)
     
    i = len(r)
    j = len(h)
    
    insert = 0
    delete = 0
    match = 0
    mismatch =0
    
    while (i>0 or j>0):
        if (d[i-1][j-1]<d[i-1][j] and d[i-1][j-1]<d[i][j-1]):
            if r[i-1] == h[j-1]:
                match += 1
            else:
                mismatch +=1               
            i = i-1
            j = j-1
       
        elif (d[i-1][j] < d[i][j-1]):
            i=i-1
            delete = delete +1
        else:
            insert = insert + 1
            j=j-1
    return insert
```
